   Curves to graph
    - y=0n+1 and label the curve "O(1)"
    - y=log(n) and label the curve "O(log n)"
    - y=n and label the curve "O(n)"
    - y=n∗log(n) and label it "O(n log n)"
    - y=n2 and label it "O(n^2)"
    - y=2n and label it "O(2^n)"
    - y=n! and label it "O(n!)"
    - y=nn and label it "O(n^n)"

```python
import math
import matplotlib.pyplot as plt
```

```python
x = range(1, 20)

# y=0n+1 and label the curve "O(1)"
y1 = [1 for n in x]
# y=log(n) and label the curve "O(log n)"
y2 = [math.log(n) for n in x]
# y=n and label the curve "O(n)"
y3 = x
# y=n∗log(n) and label it "O(n log n)"
y4 = [(n * math.log(n)) for n in x]
# y=n2 and label it "O(n^2)"
y5 = [n**2 for n in x]
# y=2n and label it "O(2^n)"
y6 = [2**n for n in x]
# y=n! and label it "O(n!)"
y7 = [math.factorial(n) for n in x]
# y=nn and label it "O(n^n)"
y8 = [n^n for n in x]
```

```python
plt.figure(figsize=(16,8))

plt.title('Big O Notation', fontsize=20)
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.plot(x, y1, color='cyan', label='$O(1)$', alpha=.8)
plt.plot(x, y2, color='red', label='$O(log n)$', alpha=.3)
plt.plot(x, y3, color='b', label='$O(n)$', alpha=.4)
plt.plot(x, y4, color='g', label='$O(n log n)$', alpha=.5)
plt.plot(x, y5, color='coral', label='$O(n^2)$', alpha=.6)
plt.plot(x, y6, color='olive', label='$O(2^n)$', alpha=.7)
plt.plot(x, y7, color='lime', label='$O(n!)$', alpha=.8)
plt.plot(x, y8, color='violet', label='$O(n^n)$', alpha=.9)

plt.xlim(1, 11)
plt.ylim(0, 50)

plt.legend(loc='upper right', fontsize=14)
plt.show()
```
