1. Use matplotlib to plot the following equation:

    y = x**2 âˆ’ x + 2

   You'll need to write the code that generates the x and y points.
   Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-10,10))
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.text(0,0, '(0,0)', fontsize=10)
plt.title('$y=x^2 - x +2$')
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

    - y=âˆšx
    - y=x3
    - y=2x
    - y=1/(x+1)
    
   You can use functions from the math module to help implement some of the equations above.

```python
x = range(1,100)

# y = sqrt(x):

y = [math.sqrt(n) for n in x]
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)
plt.show()

#  y=x3 
y = [n ** 3 for n in x]
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)
plt.show()

#   y=2x 
y = [(2 ** n) for n in x]
plt.title('$y = 2^x $')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)
plt.show()

#  $ y=1/(x+1) $
y = [(1/(n + 1)) for n in x]
plt.title('$y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(8,5))
# y = sqrt(x):
plt.subplot(221)
y = [math.sqrt(n) for n in x]
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)

# $ y=x3 $
plt.subplot(222)
y = [n ** 3 for n in x]
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)

#  $ y=2**x $
plt.subplot(223)
y = [(2 ** n) for n in x]
plt.title('$y = 2^x $')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)

#  $ y=1/(x+1) $
plt.subplot(224)
y = [(1/(n + 1)) for n in x]
plt.title('$y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y)

plt.tight_layout()
plt.suptitle('4 subplots')
plt.show()
```

```python
x = range(1,100)

# y = sqrt(x):

y = [math.sqrt(n) for n in x]
# plt.title('$y = \sqrt{x}$')
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.plot(x,y, label='$y = \sqrt{x}$')
# plt.show()

#  y=x3 
y = [n ** 3 for n in x]
# plt.title('$y = x^3$')
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.plot(x,y, label='$y = x^3$')
# plt.show()

#   y=2x 
y = [(2 ** n) for n in x]
# plt.title('$y = 2^x $')
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.plot(x,y, label='$y = 2^x $')
# plt.show()

#  $ y=1/(x+1) $
y = [(1/(n + 1)) for n in x]
# plt.title('$y = 1/(x+1)$')
plt.plot(x,y, label='$y = 1/(x+1)$', color='cyan')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Four Different Equations (Together)')
plt.ylim(0,20)
plt.xlim(1,20)
plt.legend()
plt.show()
```
