  You'll need to write the code that generates the x and y points.

```python
x = list(range(-10,10))
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.text(0,0, '(0,0)', fontsize=10)
plt.title('$y=x^2 - x +2$')
plt.show()
```
